### Generic Gradient Descent

```python
import numpy as np

data = np.loadtxt("diabetes_dataset.csv",delimiter=",")
```

```python
import pandas as pd
df = pd.DataFrame(data)

x = df.iloc[:,0:10]
x[10]=1

y = df.iloc[:,10]
```

```python
df.head()
```

```python
x.shape[1]    #no. of features
N=x.shape[0]  #no of rows
```

```python
def slope(x,y,m,k):
    m_slope=0;
    
    for j in range(x.shape[0]):
        product=0
        for i in range(x.shape[1]):
            product += m[i]*x.iloc[j,i]
        m_slope += (-2/N)*( y[j] - product )*x.iloc[j,k]
    
    return m_slope



def cost(x,y,m):
    cost=0
    for j in range(N):
        product=0
        for i in range(x.shape[1]):
            product += m[i]*x.iloc[j,i]
        cost += (1/N)*( y[j] - product)**2
    return cost

def findmc(x,y):
    m=[];
    for i in range(x.shape[1]):
        m.append(0)
    lr=0.9
    turns=1000
    for i in range(turns):
#         slopelist=slope(x,y,m)
        for j in range(len(m)):
            m[j] = m[j] - lr*slope(x,y,m,j)
        print(cost(x,y,m))
    
    print(m)
```

```python
findmc(x,y)
```
